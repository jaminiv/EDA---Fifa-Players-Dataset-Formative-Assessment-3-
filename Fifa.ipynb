```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Imported the csv file called fifa_data.csv
Fifa=pd.read_csv('fifa_data.csv')
Fifa
```

```python
display(Fifa.info()) #just to see the how many columns are there in the above table and to check the size
```

```python
Fifa.duplicated() #to check for duplicate files on this table
```

```python
#1.Which country has the most number of players (score :1)
total=Fifa['Nationality'].value_counts()
print(total)
Country = Fifa['Nationality'].value_counts().idxmax() #index max
max_count = Fifa['Nationality'].value_counts().max() #max number of times

print("\nMost number of player are from",Country,"with a total of",max_count,"people")

#Insight: In order to find the the country on which the players are from, I had to chose 'Nationality'. After filtering the table given above i found that England has the heighest number of players.
```

```python
#2.Plot a bar chart of 5 top countries with the most number of players. (score :1)
Tab = Fifa['Nationality'].value_counts()
top= Tab.head(5)
top
font_dict = {
    "family": "Serif",  # Font family name (e.g., Arial, Times New Roman)
    "weight": "bold",   # Font weight (e.g., normal, bold)
    "size": 12         # Font size in points
}
top.plot(kind='bar',color='maroon')
plt.title('Top 5 countries wiht Most players',fontdict=font_dict)
plt.xlabel('Country',fontdict=font_dict)
plt.ylabel('Number of Players',fontdict=font_dict)
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()

#Insight: The top 5 countries with most players were from England, Germany, spain, Argentina and France. The below bar graph shows clearly on the number of players from each place.
```

```python
#3.Which player has the highest salary? (score :1)
Fifa = Fifa[Fifa['Wage']!='€0']
Fifa['Wage']=Fifa['Wage'].replace('[\€]','',regex=True)
Fifa['Wage']=Fifa['Wage'][:-1].astype(float)
Maxsal=Fifa['Wage'].max()
index_sal=Fifa['Wage'].idxmax()
Playersal=Fifa.loc[index_sal,"Name"]
print("\n The highest paid Player is",Playersal,"The amount pad is €",Maxsal,"K")

#Insight: With Messi's number of fans and trophies one, there were not a doubt he would be the highest paid player. 
```

```python
#4.Plot a histogram to get the salary range of the players. (score :1)
Ht=Fifa.sort_values(by=["Wage"], ascending=False)
Ht[['Name', 'Wage']]
plt.hist(Fifa['Wage'])
plt.title('Salary Range of the Fifa Player')
plt.xlabel('Salary(k)')
plt.ylabel('Number of Players')
plt.grid(True)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
#5.Who is the tallest player in the fifa? (score :1)
Ht=Fifa.sort_values(by=["Height"], ascending=False)
Ht[['Name', 'Height']].head(2)
#Insight:Tallest player s T.Holy with height 6'9.
```

```python
#6.Which club has the most number of players? (score :1)
clubplayerc=Fifa['Club'].value_counts()
maxcount=clubplayerc.max()
clubwithmax=clubplayerc[clubplayerc==maxcount]
print("Clubs with most players ({} players each):".format(maxcount))
print(clubwithmax)
#Insight: There is max of 33 players in 23 clubs.
```

```python
#7.Which foot is most preferred by the players?Draw a bar chart for preferred foot (score :1)
footsize=Fifa['Preferred Foot'].value_counts()
print(footsize)
footsize.plot(kind='bar',color='maroon')
plt.title('Preferred foot of the player')
plt.xlabel('Foot Preference')
plt.ylabel('Number of Players')
plt.legend()
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
#Insight:Most of the players are from right foot. 
```
